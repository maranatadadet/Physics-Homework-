```python
import numpy as np

# Implementation of e**-x
print('implementation of e**(-x), all three ways ')
print(             '')
# ---------------------------------------- numpy value -----------------------------------------
print('numpy value')
for x in range(0,110,10):
    e = np.math.exp(-x)
    print(x,e)
print('                ') 
# ---------------------------------------- Inverse Case -----------------------------------------
print('inverse case')
for x in range(0,110,10):
    sn = 0
    for n in range(1000):
        sn += ((-(x**n) / np.math.factorial(n)))
    exp_x = sn
    print(x, 1/exp_x)
print('                ')   

# ---------------------------------------- Simple Case ------------------------------------------
print('simple case')
for x in range (0, 110,10):
    exp = 0
    for n in range(1000):
        exp += ( ((-1)**n) * (((x)**n) / (np.math.factorial(n)) ))
    print(x, exp)
print('             ')

# ---------------------------------------- resursion case ------------------------------------------

print('recursion case')
def sn(n,x):
    if n==0:
        return 1
    else:
        return sn(n-1,x)*(-x/n)
for x in range(0,110,10):
    expx = 0
    for n in range(1000):
        expx += sn(n,x)
    print (x, expx)
print('             ')
#----------------------------------------------------------------------------------------------------
#----------------------------------------------------------------------------------------------------

# Implement of the orbit height using function

#the altitude h above the earths surface that the satellite must have is h = (((GM(T**2))/4*(pi**2))**1/3)-R
# Need T in minutes and h in Km
# conversions:
            # 1000 m in 1 Km 
            # 60 s in 1 min

import numpy as np

T = (int(input("Please enter the satelite's period in minutes : ")))

def height(P):
    G= 6.67*(10**-11)                  # Gravitational constant 
    M= 5.97*(10**24)                   # Mass of earth 
    R= 6371                            # Radius of earth Km
    #1000*R                            # radius in m
    h = ((G*M*((P*60)**2))/(4*(np.pi**2)))**(1/3)-(1000*R)
    H = h/1000
    return H

print("Your satelite's geosynchronous height is",height(T),'Kilometers above the surface')
```
